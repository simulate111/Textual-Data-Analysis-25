```python
def extract_named_entities(article_text):
    prompt = f"extract all name entities from {article_text} and make the list for each article"
    response = openai.ChatCompletion.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}]
    )
    return response['choices'][0]['message']['content']
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
#3) Take 10 news articles from the same news data collection (Finnish or English), verify that the selected articles are not extremely long (should be less than 300 words each), discard or truncate longer documents.
df = pd.read_json('news-en-2021.jsonl', lines=True)
filtered_df = df[df['text'].apply(lambda text: len(text.split())) <= 300]
first_10_articles = filtered_df.head(10)
display(df.head())
display(first_10_articles[['text']])
```

```python
#4) Extract named entities using gpt-4o-mini model from these 10 articles, and store/print the results.
for i, article in enumerate(first_10_articles['text']):
    named_entities = extract_named_entities(article)
    print(f"Article {i+1}:")
    print(named_entities)
    print("\n")
```
