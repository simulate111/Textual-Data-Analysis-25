```python
import json
import requests
import dspy
```

```python
#Dataset
url = "http://dl.turkunlp.org/TKO_8964_2023/news-en-2021.jsonl"
response = requests.get(url)
data = response.text.splitlines()
```

```python
#First 30 news articles
articles = [json.loads(line) for line in data[:30]]
```

```python
#API key
api_key = "-"
lm = dspy.LM('openai/gpt-4o-mini', api_key=api_key)
dspy.configure(lm=lm)
```

```python
# Prompt
def extract_triples(article):
    text = article['text']
    prompt = f"Extract all instances of (person, organization, relation-to-organization) triples from the following text:\n\n{text}"
    response = lm(prompt)
    triples = response[0].strip().split('\n')
    return [tuple(triple.split(',')) for triple in triples]
```

```python
#first 30 articles
triples = []
for article in articles:
    triples.extend(extract_triples(article))
```

```python
for triple in triples:
    print(triple)
```
